```python
# Dữ liệu mẫu (thay thế bằng dữ liệu của bạn)
categories = ['A', 'B', 'C', 'D', 'E']
values = [25, 40, 30, 55, 15]

# Tạo DataFrame từ dữ liệu
data = pd.DataFrame({'Danh mục': categories, 'Giá trị': values})

# Tạo biểu đồ cột ngang bằng Seaborn
sns.barplot(x='Giá trị', y='Danh mục', data=data, orient='h')

# Thêm tiêu đề và nhãn trục
plt.title('Biểu đồ cột ngang mẫu với Seaborn')
plt.xlabel('Giá trị')
plt.ylabel('Danh mục')

# Hiển thị biểu đồ
plt.show()
```

```python
# Dữ liệu mẫu (thay thế bằng dữ liệu của bạn)
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0, 0.1, 0, 0)  # "nổi" phần thứ hai

# Vẽ biểu đồ hình tròn
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

# Đặt tiêu đề
plt.title('Biểu đồ hình tròn mẫu')

# Đảm bảo hình tròn không bị méo
plt.axis('equal')

# Hiển thị biểu đồ
plt.show()
```

```python
# Dữ liệu mẫu
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x**2
y4 = np.exp(-x)

# Tạo figure và 4 axes
fig, axes = plt.subplots(2, 2, figsize=(10, 8))  # 2 hàng, 2 cột

# Vẽ biểu đồ trên từng axes
axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Đồ thị sin(x)')

axes[0, 1].plot(x, y2)
axes[0, 1].set_title('Đồ thị cos(x)')

axes[1, 0].plot(x, y3)
axes[1, 0].set_title('Đồ thị x^2')

axes[1, 1].plot(x, y4)
axes[1, 1].set_title('Đồ thị exp(-x)')

# Điều chỉnh khoảng cách giữa các biểu đồ
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```

```python
# Dữ liệu mẫu
x = np.linspace(0, 10, 100)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = [20, 35, 30, 25]
labels = ['Phần 1', 'Phần 2', 'Phần 3', 'Phần 4']
sizes = [15, 30, 45, 10]
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

# Tạo figure và 4 axes
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Vẽ 4 kiểu biểu đồ khác nhau
axes[0, 0].plot(x, y)
axes[0, 0].set_xlabel("SDFSDF")
axes[0, 0].set_title('Đồ thị đường')

axes[0, 1].bar(categories, values)
axes[0, 1].set_title('Biểu đồ cột')

axes[1, 0].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axes[1, 0].set_title('Biểu đồ hình tròn')

axes[1, 1].scatter(x_scatter, y_scatter)
axes[1, 1].set_title('Biểu đồ phân tán')

# Điều chỉnh khoảng cách giữa các biểu đồ
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```

```python
# Dữ liệu mẫu
x = np.linspace(0, 10, 100)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = [20, 35, 30, 25]
labels = ['Phần 1', 'Phần 2', 'Phần 3', 'Phần 4']
sizes = [15, 30, 45, 10]

# Tạo figure và axes sử dụng subplot_mosaic
fig, axd = plt.subplot_mosaic([
    ['left', 'right'],
    ['bottom', 'right']
])

# Vẽ 3 kiểu biểu đồ khác nhau
axd['left'].plot(x, y)
axd['left'].set_title('Đồ thị đường')

axd['bottom'].bar(categories, values)
axd['bottom'].set_title('Biểu đồ cột')

axd['right'].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axd['right'].set_title('Biểu đồ hình tròn')

# Điều chỉnh khoảng cách giữa các biểu đồ
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```
